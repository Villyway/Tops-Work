# Section B: Practical Task

```python
import pandas as pd
import numpy as np
```

```python
# Load Dataset
df = pd.read_csv("sales_data_sample.csv", encoding="latin1")

print("Dataset Loaded Successfully")
print("Dataset Shape:", df.shape)

df.head()
```

## Task 1: Clean PRODUCTLINE Column
Remove extra spaces and apply title casing.

```python
# Clean PRODUCTLINE
df['PRODUCTLINE'] = df['PRODUCTLINE'].str.strip().str.title()

print("PRODUCTLINE Column Cleaned Successfully")

df[['PRODUCTLINE']].head()
```

## Task 2: Convert SALES Column into Float Data Type

```python
# Convert SALES
df['SALES'] = pd.to_numeric(df['SALES'], errors='coerce')

print("SALES Data Type:")
print(df['SALES'].dtype)
```

# Task 3: Create Dictionary for Product Line Wise Sales

```python
# Create Dictionary
sales_dict = {}

for index, row in df.iterrows():

    product = row['PRODUCTLINE']
    sales = row['SALES']

    if pd.isna(sales):
        continue

    if product in sales_dict:
        sales_dict[product] += sales
    else:
        sales_dict[product] = sales

print("Dictionary Created Successfully")
```

```python
# Display Dictionary
sales_dict
```

# Task 4: Categorize Product Lines Based on Revenue

```python
# Revenue Classification
for product, total in sales_dict.items():

    if total > 100000:
        print(product, "-> High Revenue Product")
    else:
        print(product, "-> Normal Revenue Product")
```

# Title:
Automated Sales Integrity & Reporting Engine

# Problem Statement

The goal of this project is to improve the quality of the retail sales dataset by identifying data quality issues, validating sales data, detecting revenue outliers, and generating an automated summary report.

The project helps ensure that business decisions are based on clean and reliable sales information.

# Step 1: Load Dataset

```python
df = pd.read_csv("D:\sales_data_sample.csv", encoding="latin1")

print("Dataset Loaded Successfully")
print("Rows:", df.shape[0])
print("Columns:", df.shape[1])
```

# Step 2: Data Cleaning
Standardize PRODUCTLINE values and convert SALES into numeric format.

```python
df['PRODUCTLINE'] = df['PRODUCTLINE'].str.strip().str.title()

df['SALES'] = pd.to_numeric(
    df['SALES'],
    errors='coerce'
)

print("Data Cleaning Completed")
```

# Step 3: Check Missing Values

```python
missing_values = df.isnull().sum()

print("Missing Values Per Column")
print(missing_values)
```

# Step 4: Identify Data Type Issues

```python
print("Column Data Types")

print(df.dtypes)
```

# Step 5: Revenue Outlier Detection
Outliers are detected using the Mean + 2 × Standard Deviation method.

```python
mean_sales = df['SALES'].mean()

std_sales = df['SALES'].std()

outlier_threshold = mean_sales + (2 * std_sales)

outliers = df[
    df['SALES'] > outlier_threshold
]

print("Outlier Threshold:")
print(round(outlier_threshold, 2))

print("\nNumber of Outliers:")
print(len(outliers))
```

# Step 6: Display Revenue Outliers

```python
outliers[
    [
        'ORDERNUMBER',
        'PRODUCTLINE',
        'SALES'
    ]
].head(10)
```

# Step 7: Automated Console Summary Report

```python
print("=" * 50)

print("AUTOMATED SALES INTEGRITY REPORT")

print("=" * 50)

print("Total Records :", len(df))

print("Total Columns :", len(df.columns))

print("Missing Values :",
      df.isnull().sum().sum())

print("Average Sales :",
      round(df['SALES'].mean(), 2))

print("Maximum Sales :",
      round(df['SALES'].max(), 2))

print("Minimum Sales :",
      round(df['SALES'].min(), 2))

print("Revenue Outliers Found :",
      len(outliers))

print("=" * 50)

print("Report Generated Successfully")

print("=" * 50)
```
